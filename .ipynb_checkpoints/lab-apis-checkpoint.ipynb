## Exercise 1: Discovering New Music through Your Favorite Artists

**Objective:** 
Uncover new music by exploring the top tracks of your favorite artists and their related artists.

**Instructions:**

```python



# Fetch Top Tracks
artists = ["Sabrina Carpenter", "Chappell Roan", "Ariana Grande"]
top_songs = {}

def get_top_tracks(artists):
    for artist in artists:
        print(f'Searching for artist: {artist}')
        
        # Search for the artist and get the artist ID
        result = sp.search(q=artist, type='artist', limit=1)
        artist_id = result["artists"]["items"][0]['id']
        print(f'Artist ID for {artist} is {artist_id}')
        
        # Get the top tracks for the artist using their ID
        top_tracks = sp.artist_top_tracks(artist_id)  
        #print(f'Top tracks response for {artist}: {top_tracks}')
        
        # Extract the top 5 track names
        top_track_names = [track['name'] for track in top_tracks['tracks'][:5]]
        print(f'{artist} top tracks are: {top_track_names}')
        
        # Store the top tracks in the dictionary
        top_songs[artist] = top_track_names
    
    return top_songs

# Call the function to get the top tracks for the artists
top_songs = get_top_tracks(artists)
print(top_songs)
```

```python
#Discover Related Artists:

#Write a function named find_related_artists.
#This function should accept an artist's name and return the names of the first 5 artists related to the provided artist.
#Store the results in a list named related_artists_list.
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Initialize Spotify client
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="your-client-id", client_secret="your-client-secret"))

def find_related_artists(artist_name):
    # Search for the artist by name
    result = sp.search(q=artist_name, type='artist', limit=1)
    artist_id = result['artists']['items'][0]['id']  # Get the artist's ID
    
    # Get related artists using the recommendations endpoint
    recommendations = sp.recommendations(seed_artists=[artist_id], limit=5)
    
    # Extract the first 5 related artist names
    related_artists_names = [artist['name'] for artist in recommendations['artists']]
    
    return related_artists_names

# Example usage
artist_name = "Sabrina Carpenter"
related_artists_list = find_related_artists(artist_name)
print(f"Related artists for {artist_name}:")
print(related_artists_list)
```

To fetch details about a specific playlist, you can use the playlist method. You'll need the playlist's Spotify ID.

In this example, we will use the following playlist id: *37i9dQZF1DXd9zR7tdziuQ*

```python
playlist_id = "37i9dQZF1DXd9zR7tdziuQ"
playlist = sp.playlist(playlist_id)
```

```python
print(playlist['name'])  # Print the playlist's name
print(playlist['description'])  # Print the playlist's description
```

### Getting Tracks from a Playlist
If you want to get the tracks from a specific playlist, you can use the playlist_tracks method.

```python
tracks = sp.playlist_tracks(playlist_id)
for track in tracks['items']:
    print(track['track']['name'])  # Print each track's name
```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        if artist_name not in artists_list:  # This ensures each artist is added only once
            artists_list.append(artist_name)

print(artists_list)
```

## Exercise 2: Unraveling the World of Playlists


1. **Featured Exploration**: 
   - Fetch the list of Spotify's current featured playlists. 
   - Extract and display the names and IDs of the top 5 featured playlists.
   
2. **Deep Dive**:
   - Choose any one of the top 5 featured playlists (you can choose the one you personally find most interesting or simply pick one randomly).
   - Fetch and display its name, description, and total track count.

3. **Track-tastic**:
   - Extract and display the names of the first 10 tracks in the chosen playlist.

4. **Artistic Flair**:
   - Create a dictionary where the keys are the names of the first 10 tracks, and the values are lists containing the names of the artists associated with each track.
   - For example: `{"TrackName1": ["Artist1", "Artist2"], "TrackName2": ["Artist3"]}`
   

```python
# Your answer here
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Initialize Spotify client
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="your-client-id", client_secret="your-client-secret"))

# 1. Fetch the list of featured playlists
featured_playlists = sp.featured_playlists(country='US', limit=5)

# 2. Extract and display the names and IDs of the top 5 featured playlists
top_5_playlists = []
for playlist in featured_playlists['playlists']['items'][:5]:
    name = playlist['name']
    playlist_id = playlist['id']
    top_5_playlists.append((name, playlist_id))

print("Top 5 Featured Playlists:")
for i, (name, playlist_id) in enumerate(top_5_playlists, 1):
    print(f"{i}. {name} (ID: {playlist_id})")

# Choose one of the top 5 playlists, for example, the first one
chosen_playlist_id = top_5_playlists[0][1]

# 3. Fetch and display the name, description, and total track count of the chosen playlist
chosen_playlist = sp.playlist(chosen_playlist_id)
playlist_name = chosen_playlist['name']
playlist_description = chosen_playlist['description']
track_count = chosen_playlist['tracks']['total']

print(f"\nChosen Playlist - {playlist_name}:")
print(f"Description: {playlist_description}")
print(f"Total Tracks: {track_count}")

# 4. Extract and display the names of the first 10 tracks in the chosen playlist
tracks = sp.playlist_tracks(chosen_playlist_id, limit=10)
track_names = [track['track']['name'] for track in tracks['items']]

print("\nFirst 10 Tracks in the Playlist:")
for i, track_name in enumerate(track_names, 1):
    print(f"{i}. {track_name}")

# 5. Create a dictionary where track names are keys and the associated artists are values
track_artist_dict = {}

for track in tracks['items']:
    track_name = track['track']['name']
    artists = [artist['name'] for artist in track['track']['artists']]
    track_artist_dict[track_name] = artists

print("\nTrack-Artist Dictionary:")
print(track_artist_dict)
```
